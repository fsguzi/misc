```python
#3-way capable
def within_tri_twin(_c,_g):   
    global incentive,incentive_g,assignment_c,assignment_g
    
    tot_incentive = incentive[_c].sum(axis=1)
    unique_g = np.unique(_g)
    num_unique_g = len(unique_g)

    num_single_needed = _c.shape[1]

    out_score = np.zeros((num_unique_g,1000),dtype = 'int')
    out_ind = np.zeros((num_unique_g,1000),dtype = 'int')
    for g_ind,g in enumerate(unique_g):
        kids = np.where(_g==g)
        all_d_score = tot_incentive[kids]

        best_score_ind = all_d_score.argmax(axis=0)
        best_score = all_d_score[best_score_ind,np.arange(1000)]
        best_sib = kids[0][best_score_ind]
        out_score[g_ind,:] = best_score
        out_ind[g_ind,:] = best_sib

    out_score = out_score[:,unique_g]
    out_ind = out_ind[:,unique_g]

    giftind0 = np.repeat(np.arange(num_unique_g),num_unique_g)
    giftind2 = np.tile(np.arange(num_unique_g),num_unique_g)

    cubic = np.tile(out_score.reshape((num_unique_g,num_unique_g,1)),(1,1,num_unique_g))+out_score.T[giftind0.reshape((num_unique_g,num_unique_g))] + out_score[giftind2.reshape((num_unique_g,num_unique_g))]

    total = num_unique_g*num_unique_g
    bestindice = cubic.argmax(axis=2)
    out_score = cubic[giftind0,giftind2,bestindice.reshape(total,)].reshape((num_unique_g,num_unique_g))

    positiveindex = np.where((out_score>0))

    if len(positiveindex[0])>0:

        improvement_record = []

        positiveindex = np.array(positiveindex).T

        for r,c in positiveindex:
            g1 = r
            g2 = c
            g3 = bestindice[r,c]
            sib1 = out_ind[g1,g2]
            sib2 = out_ind[g2,g3]
            sib3 = out_ind[g3,g1]
            
            score = out_score[r,c]
            first = _c[sib1]
            second = _c[sib2]
            third = _c[sib3]
            
            gift1,gift2,gift3 = unique_g[[g1,g2,g3]]
            gift = np.array([gift1,gift2,gift3])
            improvement_record.append([score,first,second,third,gift])
        return improvement_record
    else:
        return []
```

## 3<->2

```python
def between_tri_twin():    
    global incentive,incentive_g,assignment_c,assignment_g,triplet_c,triplet_g,twin_c,twin_g
    
    triplet_supply_g = np.where(np.bincount(triplet_g)>=2)[0]
    twin_supply_g = np.where(np.bincount(twin_g)>=3)[0]
    
    inter_score = np.zeros((len(triplet_supply_g),len(twin_supply_g)))
    #inter_score_g = np.zeros((len(triplet_supply_g),len(twin_supply_g)))
    players = np.zeros((len(triplet_supply_g),len(twin_supply_g),12),dtype='int')
    
    for o,o_g in enumerate(triplet_supply_g):
        for d,d_g in enumerate(twin_supply_g):
            tri_in_game = triplet_c[np.where(triplet_g == o_g)]
            twin_in_game = twin_c[np.where(twin_g == d_g)]

            tri_score = incentive[tri_in_game,d_g].sum(axis=1)
            twin_score = incentive[twin_in_game,o_g].sum(axis=1)

            #tri_score_g = incentive_g[tri_in_game,d_g].sum(axis=1)
            #twin_score_g = incentive_g[twin_in_game,o_g].sum(axis=1)

            tri_best = tri_score.argsort()[-2:]
            tri_score_best = tri_score[tri_best].sum()
            twin_best = twin_score.argsort()[-3:]
            twin_score_best = twin_score[twin_best].sum()

            
            #tri_score_g_best = tri_score_g[tri_best].sum()
            
            #twin_score_g_best = twin_score_g[twin_best].sum()

            inter_score[o,d] = tri_score_best+twin_score_best
            #inter_score_g[o,d] = tri_score_g_best+twin_score_g_best
            players[o,d] = np.concatenate([tri_in_game[tri_best].ravel(),twin_in_game[twin_best].ravel()])
    
    where_is_improvement = np.where((inter_score>0))#  & (inter_score+inter_score_g>0))
    
    if len(where_is_improvement[0])>0:
        improvement_record = []
        
        swap_order = np.array(where_is_improvement).T
        
        for r,c in swap_order:
            score = inter_score[r,c]
            triplets_num = players[r,c,:6]         
            twins_num = players[r,c,6:]
            
            g1 = triplet_supply_g[r]
            g2 = twin_supply_g[c]
            
            gift = np.array([g1,g1,g2])
            improvement_record.append([score,
                                      triplets_num,
                                      triplets_num,
                                      twins_num,
                                      gift])
        return improvement_record
    else:
        return []
```

## 3/2 <-> 1

```python
def sin_tri_twin(_c,_g):  
    global incentive,incentive_g,assignment_c,assignment_g
    
    tot_incentive = incentive[_c].sum(axis=1)
    unique_g = np.unique(_g)
    num_unique_g = len(unique_g)

    num_single_needed = _c.shape[1]

    out = np.zeros((num_unique_g,1000,2),dtype = 'int')
    for g_ind,g in enumerate(unique_g):
        kids = np.where(_g==g)
        all_d_score = tot_incentive[kids]

        best_score_ind = all_d_score.argmax(axis=0)
        best_score = all_d_score[best_score_ind,np.arange(1000)]
        best_sib = kids[0][best_score_ind]
        out[g_ind,:,0] = best_score
        out[g_ind,:,1] = best_sib


    giftind0 = np.repeat(np.arange(1000),num_unique_g)
    giftind2 = np.tile(np.arange(num_unique_g),1000)
    where_non_single = np.where(assignment_g<45001)
    cube = incentive[assignment_g][:,:,unique_g]

    #make incetive from non-single bad
    cube[where_non_single] = -9999

    best_score_ind = cube.argsort(axis=1)[:,-num_single_needed:,:]
    inc_score = np.zeros((num_unique_g,1000),dtype='int')
    best_score_who = []
    for n in range(num_single_needed):                   
        inc_score += cube[giftind0,best_score_ind[:,n,:].ravel(),giftind2].reshape((1000,num_unique_g)).T   
        #num_unique_g * 1000
        best_score_who.append(assignment_g[giftind0,best_score_ind[:,n,:].ravel()].reshape((1000,num_unique_g)).T)
        
    best_score_who = np.stack(best_score_who,axis = 2)
    
    out[:,:,0] += inc_score

    positiveindex = np.where((out[:,:,0]>0))
    
    if len(positiveindex[0])>0:
        
        improvement_record = []
        
        positiveindex = np.array(positiveindex).T
        
        for r,c in positiveindex:
            score = out[r,c,0]
            first = _c[out[r,c,1]]
            second = first
            third = best_score_who[r,c]
            
            g1 = unique_g[r]
            
            
            gift = np.array([g1,g1,c])
            improvement_record.append([score,first,second,third,gift])
        return improvement_record
    else:
        return []
```

```python
def within_single():    
    global incentive,assignment_c,assignment_g
    
    #construct all of swap matrix    
    giftind0 = np.repeat(np.arange(1000),1000)
    giftind2 = np.tile(np.arange(1000),1000)
    cubic = incentive[assignment_g]
    
    #make swapping with non-single bad
    where_are_non_single = np.where(assignment_g<45001)
    cubic[where_are_non_single] = -9999
    
    outgoingindice = cubic.argmax(axis=1)
    outgoing = cubic[giftind0,outgoingindice.reshape(1000000,),giftind2].reshape((1000,1000))
    cubic = np.tile(outgoing.reshape((1000,1000,1)),(1,1,1000))+outgoing.T[giftind0.reshape((1000,1000))] + outgoing[giftind2.reshape((1000,1000))]

    bestindice = cubic.argmax(axis=2)
    swapscore = cubic[giftind0,giftind2,bestindice.reshape(1000000,)].reshape((1000,1000))

    best1st = assignment_g[giftind0,outgoingindice.reshape(1000000,)].reshape((1000,1000))


    best2nd_ind = outgoingindice[giftind2,bestindice.reshape(1000000,)]
    best2nd = assignment_g[giftind2,best2nd_ind].reshape((1000,1000))


    best3rd_ind = outgoingindice[bestindice.reshape(1000000,),giftind0]
    best3rd = assignment_g[bestindice.reshape(1000000,),best3rd_ind].reshape((1000,1000))
    
    positiveindex = np.where((swapscore>0))# & (best1st>=45001) & (best2nd>=45001) & (best3rd>=45001))  
    
    if len(positiveindex[0])>0:
        
        improvement_record = []
        
        positiveindex = np.array(positiveindex).T
        
        for r,c in positiveindex:
            score = swapscore[r,c]
            first = np.array([best1st[r,c],])
            second = np.array([best2nd[r,c],])
            third = np.array([best3rd[r,c],])
            
            gift = np.array([r,c,bestindice[r,c]])
            improvement_record.append([score,first,second,third,gift])
        return improvement_record
    else:
        return []
```

```python
def all_hell():
    global incentive,incentive_g,assignment_c,assignment_g,triplet_c,triplet_g,twin_c,twin_g
    def asym(_c,_g):
        tot_incentive = incentive[_c].sum(axis=1)
        unique_g = np.unique(_g)
        num_unique_g = len(unique_g)

        num_single_needed = _c.shape[1]

        out_score = np.zeros((num_unique_g,1000),dtype = 'int')
        out_ind = np.zeros((num_unique_g,1000),dtype = 'int')
        for g_ind,g in enumerate(unique_g):
            kids = np.where(_g==g)
            all_d_score = tot_incentive[kids]

            best_score_ind = all_d_score.argmax(axis=0)
            best_score = all_d_score[best_score_ind,np.arange(1000)]
            best_sib = kids[0][best_score_ind]
            out_score[g_ind,:] = best_score
            out_ind[g_ind,:] = best_sib
        return out_score,out_ind,unique_g

    out_score3,out_ind3,unique_g3 = asym(triplet_c,triplet_g)
    out_score2,out_ind2,unique_g2 = asym(twin_c,twin_g)

    out_score3 = out_score3[:,unique_g2]
    out_ind3 = out_ind3[:,unique_g2]

    out_score2 = out_score2[:,unique_g3]
    out_ind2 = out_ind2[:,unique_g3]

    cubic = incentive[assignment_g]

    #make swapping from non-single bad
    where_are_non_single = np.where(assignment_g<45001)
    cubic[where_are_non_single] = -9999

    cubic=cubic[unique_g2][:,:,unique_g3]

    total_score = out_score3 + out_score2.T + cubic.max(axis=1).T

    single_ind = cubic.argmax(axis=1)

    positiveindex = np.where(total_score>0)

    if len(positiveindex[0])>0:

        improvement_record = []

        positiveindex = np.array(positiveindex).T

        for r,c in positiveindex:
            score = total_score[r,c]
            
            g3 = unique_g3[r]
            g2 = unique_g2[c]

            c3_ = out_ind3[r,c]
            c2_ = out_ind2[c,r]

            single_ = single_ind[c,r]
            c1 = assignment_g[g2,single_]

            first = triplet_c[c3_]
            second = first
            third = np.append(twin_c[c2_],c1)
            
            gift = np.array([g3,g3,g2])
            improvement_record.append([score,first,second,third,gift])
        return improvement_record
    else:
        return []
```
